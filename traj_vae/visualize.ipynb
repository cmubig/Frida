```python
import torch
import matplotlib.pyplot as plt
from traj_helpers import plot_trajectory
```

```python
model = torch.load("saved_models/model.pt")
model.eval()
```

```python
# Visualize generated trajectories
generated_trajectories = model.sample_trajectories(100)
fig, ax = plt.subplots()
ax.set_xlim([-0.125, 0.25])
ax.set_ylim([-0.125, 0.125])
print(len(generated_trajectories))
for trajectory in generated_trajectories:
    plot_trajectory(ax, trajectory)
```
